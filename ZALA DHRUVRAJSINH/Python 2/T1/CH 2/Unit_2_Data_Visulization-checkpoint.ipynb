```python
data={"Year":[2015,2016,2017,2018,2019,2020],'Sales':[100,120,150,170,200,220],'Expences':[70,80,90,100,110,120]}
df=pd.DataFrame(data)
plt.stackplot(df['Year'],df['Sales'],df['Expences'],labels=['Sales','Expences'])
plt.legend(loc="upper left")
plt.show()
```

```python
months=range(1,13)
sales=np.array([10,15,20,30,20,50,30,70,80,40,100,10])
expences=np.array([5,10,15,20,25,30,35,40,45,50,55,60])
plt.plot(months,sales,color='green',label='sales')
plt.plot(months,expences,color='red',label='expences')
plt.fill_between(months,sales,expences,where=sales>expences,alpha=0.25,color='green',interpolate=True)
plt.fill_between(months,sales,expences,where=sales<expences,alpha=0.25,color='red',interpolate=True)
plt.legend()
plt.show()
```

```python
import requests
from wordcloud import WordCloud, STOPWORDS
url='https://www.gutenberg.org/files/11/11-0.txt'
response=requests.get(url)
text=response.text
```

```python
wc=WordCloud(background_color='white',max_words=2000000000)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
