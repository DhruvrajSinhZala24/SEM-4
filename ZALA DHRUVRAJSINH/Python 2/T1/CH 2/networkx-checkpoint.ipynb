```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g=nx.Graph()
g.add_node("A")
g.add_node("B")
g.add_node("C")
g.add_edge("A","B")
g.add_edge("C","B")
g.add_edge("A","C")
g.add_edge("A","A")
nx.draw_networkx(g)
plt.show()
```

```python
'''Question:- The following dictionary shows you how 5 people follow each other on instagram create a directed
network x graph for the given insta followers:-
instagram={'person1':[0,1,1,0,1],'person2':[0,0,1,0,1],'person2':[1,1,0,1,1],'person4':[1,1,1,0,0],'person5':[1,1,0,0,0]}
'''
instagram={'person1':[0,1,1,0,1],'person2':[0,0,1,0,1],'person3':[1,1,0,1,1],'person4':[1,1,1,0,0],'person5':[1,1,0,0,0]}
g=nx.DiGraph()
g.add_nodes_from([1,2,3,4,5])
count=1
l=list()
for i in instagram.values():
    for j in range(len(i)):
        if(i[j]==1):
            tup=(count,j+1)
            l.append(tup)
    count+=1
print(l)
g.add_edges_from(l)
nx.draw_networkx(g)
plt.show()
```

```python
# directly above question without loop
import numpy as np 
G=nx.DiGraph(np.array(list(instagram.values())))
l={0:'person1',1:'person2',2:'person3',3:'person4',4:'person5'}
nx.draw_networkx(G,labels=l)
plt.show()
```
