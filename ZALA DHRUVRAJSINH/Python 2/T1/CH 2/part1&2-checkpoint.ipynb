```python
import pandas as pd
df=pd.read_csv("supermarket_sales.csv")
df
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['gross income'].fillna(df['gross income'].mean(),inplace=True)
df.info()
```

```python
df['customer type'].fillna(df['customer type'].mode()[0],inplace=True)
df['gender'].fillna(df['gender'].mode()[0],inplace=True)
df.info()
```

```python
x=df.loc[df["branch"]=='A','total'].sum()
y=df.loc[df["branch"]=='B','total'].sum()
z=df.loc[df["branch"]=='C','total'].sum()
print(f"Revenue of branch a:- {x}")
print(f"Revenue of branch b:- {y}")
print(f"Revenue of branch c:- {z}")
```

```python
# df['product line'].value_counts()
d={}
for i in df['product line']:
    d[i]=d.get(i,0)+1
d=sorted(d.items(), key=lambda x:x[1], reverse=True)[:5]
print(d)
```

```python
rating_a=df.loc[df['branch']=='A','rating'].mean()
rating_b=df.loc[df['branch']=='B','rating'].mean()
rating_c=df.loc[df['branch']=='C','rating'].mean()
print(f"Rating of branch A:- {rating_a}")
print(f"Rating of branch B:- {rating_b}")
print(f"Rating of branch C:- {rating_c}")
```

```python
d={}
for i in df['payment']:
    d[i]=d.get(i,0)+1
d=sorted(d.items(), key=lambda x:x[1], reverse=True)
print(d[0])
```

```python
total=df.loc[df['customer type']=='Member','customer type'].count()
print(total)
```

```python
pd.crosstab(df['customer type'],df['payment'])
```

```python
week=sum(df.loc[(df['day_of_week']=='Saturday')| (df['day_of_week']=='Sunday'),'total'])
print(week)
```

```python
non_week=sum(df.loc[(df['day_of_week']!='Saturday')&(df['day_of_week']!='Sunday'),'total'])
print(non_week)
```
