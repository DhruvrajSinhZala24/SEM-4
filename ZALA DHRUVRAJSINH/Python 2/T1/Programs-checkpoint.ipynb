```python
import matplotlib.pyplot as plt
df=pd.read_csv("supermarket_sales.csv")
print(df.head(8))
print(df.info())
print("--------------------------------------")
# df1=df.fillna(df.mean())
# print(df1.info())
df["cogs"].fillna(df["cogs"].mean(),inplace=True)
df["Rating"].fillna(df["Rating"].mean(),inplace=True)
print(df.info())
print("--------------------------------------")
print(df.loc[(df["Quantity"]<3) & ((df["Rating"]>=8.5) | (df["Total"]>600))])
print("--------------------------------------")
print("Members Total:- ",df.loc[df["Customer type"]=="Member"]["Total"].sum())
print("Normal Customers Total:- ",df.loc[df["Customer type"]=="Normal"]["Total"].sum())
print("--------------------------------------")
Ewallet=df.loc[df["Payment"]=="Ewallet"]["Total"].sum()
Cash=df.loc[df["Payment"]=="Cash"]["Total"].sum()
Credit_card=df.loc[df["Payment"]=="Credit card"]["Total"].sum()
plt.pie([Ewallet,Cash,Credit_card],autopct="%0.1f%%",labels=["Ewallet","Cash","Credit Card"])
plt.show()
print("--------------------------------------")
print("Male:- ",df.loc[df["Gender"]=="Male"]["Total"].mean())
print("Female:- ",df.loc[df["Gender"]=="Female"]["Total"].mean())
print("--------------------------------------")
plt.scatter(df['Total'],df['Rating'])
plt.show()
print("--------------------------------------")
fig,ax=plt.subplots()
ax.boxplot([df['Rating'],df['Quantity']])
plt.show()
print("There are no outliers")
print("--------------------------------------")
p = pd.plotting.parallel_coordinates(df,'Product line',cols=[ 'Unit price', 'Total', 'cogs'],color=['red','blue','green','yellow','orange'])
plt.show()
```
