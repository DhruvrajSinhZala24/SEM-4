```python
df.loc[df['horsepower']=='?','horsepower'].count()
df=df.loc[df['horsepower']!='?']
# count=0
# for i in x:
#     if i=="?":
#         count+=1
# print(count)
```

```python
df['horsepower']=df['horsepower'].astype("int32")
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots()
ax.boxplot([df['mpg'],df['horsepower'],df['acceleration']])
plt.show()
```

```python
q1=df['horsepower'].quantile(0.25)
q3=df['horsepower'].quantile(0.75)
iqr=q3-q1
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
print(upper_limit)
print(lower_limit)
```

```python
def function(df,col):
    d1=df[col].quantile(0.25)
    d2=df[col].quantile(0.75)
    iqr=d2-d1
    upper_limit=d1+1.5*iqr
    lower_limit=d2-1.5*iqr
    df=df.loc[df[col]<=upper_limit]
    df=df.loc[df[col]>=lower_limit]
    return df
df=function(df,'horsepower')
```

```python
fig,ax=plt.subplots()
ax.boxplot([df['mpg'],df['horsepower'],df['acceleration']])
plt.show()
```

```python
df=function(df,'acceleration')
```
