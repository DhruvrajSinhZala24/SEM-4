```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
df=pd.DataFrame(columns=["Team","Match","Won","Lost","Tied","NR","Pts","NRR"])
response=requests.get("https://www.cricbuzz.com/cricket-series/9237/indian-premier-league-2025/points-table")
content=response.text
soup=BeautifulSoup(content,"html.parser")
name=soup.find_all("div",class_="cb-col cb-col-84")
Team=[]
for i in name:
  Team.append(i.text)
df["Team"]=Team
x=soup.find_all("td",class_="cb-srs-pnts-td")
score=[]
for i in range(0,len(x)):
  y=[]
  for j in range(0,10):
    y.append(x[i].text)
  score.append(y)
df["Match"]=score[0]
df["Won"]=score[1]
df["Lost"]=score[2]
df["Tied"]=score[3]
df["NR"]=score[4]
df["Pts"]=score[5]
df["NRR"]=score[6]
df
```

```python
#Short:-
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.cricbuzz.com/cricket-series/9237/indian-premier-league-2025/points-table")
content = response.text
soup = BeautifulSoup(content, "html.parser")

name = soup.find_all('div',class_="cb-col cb-col-84")
x = soup.find_all("td", class_="cb-srs-pnts-td")

Mat = x[0::7]
Won = x[1::7]
lost = x[2::7]
tied = x[3::7]
NR = x[4::7]
pts = x[5::7]
nrr = x[6::7]

d = {"Name":[],'Mat':[],'won':[],'lost':[],'tied':[], 'NR':[],'pts':[],'nrr':[]}

for i in range(len(name)):
  d['Name'].append(name[i].text.strip())
  d['Mat'].append(Mat[i].text.strip())
  d['won'].append(Won[i].text.strip())
  d['lost'].append(lost[i].text.strip())
  d['tied'].append(tied[i].text.strip())
  d['NR'].append(NR[i].text.strip())
  d['pts'].append(pts[i].text.strip())
  d['nrr'].append(nrr[i].text.strip())
import pandas as pd
df = pd.DataFrame(d)
df
```
