```python
import socket
#not related to webscrapping it is diffrent topic
mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
mysock.connect(('example.com', 80))
# cmd = 'GET / HTTP/1.0\r\nHost: example.com\r\n\r\n'.encode()
cmd='GET http://httpbin.org/ HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)
while True:
  data = mysock.recv(512)
  if not data:
    break
  print(data.decode(), end='')
mysock.close()
```

```python
# main task
import requests
from bs4 import BeautifulSoup
response=requests.get("https://subslikescript.com/movies")
content=response.text
soup=BeautifulSoup(content,"html.parser")
box=soup.find('ul',class_="scripts-list")
l=box.find_all('a',href=True)#Only find a tags with href not neccesary for this
links=[]
for i in l:
  links.append(i['href'])
# print(links)
mainl="https://subslikescript.com"
d={}
for i in links:
  try:
    url=mainl+i
    response=requests.get(url)
    content=response.text
    soup=BeautifulSoup(content,"html.parser")
    x=soup.find('h1').text
    y=soup.find('div',class_="full-script").text.strip()
    with open(f"{x}.txt",'w',encoding='utf-8') as f:
      f.write(y)
  except Exception as e:
    print(e)
```
