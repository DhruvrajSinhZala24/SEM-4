```python
from bs4 import BeautifulSoup
import pandas as pd
f=open("indeed jobs.html",encoding="utf-8")
soup=BeautifulSoup(f,"html.parser")
details=soup.find_all("td",class_="resultContent css-1o6lhys eu4oa1w0")
d={'name-of-the-job':[],'name-of-the-company':[],'job-location':[],'job-details':[]}
for i in details:
    post=i.find('div',class_="css-pt3vth e37uo190")
    d['name-of-the-job'].append(post.text)
    company=i.find('div',class_="css-1afmp4o e37uo190")
    d['name-of-the-company'].append(company.text)
    loc=i.find('div',class_="css-1restlb eu4oa1w0") #OR WE CAN PASS DICTIONARY: 'DIV',{"class":"xyz","id":"abc"}
    d['job-location'].append(company.text)
    detail=i.find('div',class_="jobMetaDataGroup css-qspwa8 eu4oa1w0")
    d['job-details'].append(detail.text)
df=pd.DataFrame(d)
df
```
