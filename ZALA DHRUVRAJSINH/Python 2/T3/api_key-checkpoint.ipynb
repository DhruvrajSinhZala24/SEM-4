```python
import json
import requests
api='f09bb7e6c42bbe73719ec72a369ab02f'
url=f"http://api.openweathermap.org/geo/1.0/direct?q=Ahmedabad&appid={api}"
response=requests.get(url)
result=response.json()
print(json.dumps(result,indent=3))
```

```python
lat=result[0]['lat']
lon=result[0]['lon']
lat,lon
```

```python
#2
url=f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&units=metric&appid={api}"
response=requests.get(url)
result=response.json()
print(json.dumps(result,indent=3))
```

```python
#print description,temprature,pressure,wind speed,visibility
print("Description:",result['weather'][0]['description'])
print("Temprature:",result['main']['temp'],'(C)')
print("Pressure:",result['main']['pressure'])
print("Wind Speed:",result['wind']['speed'])
print("Visibility:",result['visibility'])
```

```python
#3
url=f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api}'
response=requests.get(url)
result=response.json()
print(json.dumps(result,indent=3))
```

```python
#4
url=f'http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api}'
response=requests.get(url)
result=response.json()
print(json.dumps(result,indent=3))
```

```python
import pandas as pd
d = {'Datetime':[],'temp':[],'humidity':[],'description':[]}
p = result['list']
for i in p:
    d['Datetime'].append(i['dt_txt'])
    d['temp'].append(i['main']['temp'])
    d['humidity'].append(i['main']['humidity'])
    d['description'].append(i['weather'][0]['description'])
df = pd.DataFrame(d)
df
```
