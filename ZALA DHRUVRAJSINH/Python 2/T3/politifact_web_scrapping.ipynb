```python
#1. news statement
#2. date
#3. name of author / sources
#4. label -> true / false / pants on fire / half true
import requests
from bs4 import BeautifulSoup
import pandas as pd
df=pd.DataFrame(columns=["Author","Statement","Date","Label"])
n=int(input("Enter The Number Of Pages To visit: "))
links=[]
for i in range(1,n+1):
  links.append(f"https://www.politifact.com/factchecks/list/?page={i}")
print(links)
al=[]
sl=[]
dl=[]
ll=[]
for i in links:
    print(i)
    url=i
    response=requests.get(url)
    content=response.text
    soup=BeautifulSoup(content,"html.parser")
    Author=soup.find_all('a',class_="m-statement__name")

    for j in Author:
      al.append(j.text.strip())
    Statement=soup.find_all('div',class_="m-statement__quote")

    for j in Statement:
      sl.append(j.text.strip())
    Date=soup.find_all('footer',class_="m-statement__footer")

    for j in Date:
      dl.append(j.text.split("•")[1].strip())
    Label=soup.find_all('img',class_="c-image__original",alt=True)

    for j in Label:
      ll.append(j['alt'])
df['Author']=al
df['Statement']=sl
df['Date']=dl
df['Label']=ll
print(df)
```
