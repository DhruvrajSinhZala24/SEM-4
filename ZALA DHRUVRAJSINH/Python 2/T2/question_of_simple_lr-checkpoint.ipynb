```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
df=pd.read_csv("Marksheet.csv")
df=df.loc[(df['TEST-3 MARKS']!='AB')&(df['TEST-4 MARKS']!='AB')]
df['TEST-3 MARKS']=df['TEST-3 MARKS'].astype('float64')
df['TEST-4 MARKS']=df['TEST-4 MARKS'].astype('float64')
df.info()
df['TOTAL']=df['TEST-1 MARKS']+df['TEST-2 MARKS']+df['TEST-3 MARKS']+df['TEST-4 MARKS']
df
```

```python
l=list()
for i in df['TOTAL']:
    grade=""
    if(i>=85):
        grade='A'
    elif(i>=75 and i<=84.99):
        grade='B'
    elif(i>=65 and i<=74.99):
        grade='C'
    elif(i>=50 and i<=64.99):
        grade='D'
    elif(i>=35 and i<=49.99):
        grade='E'
    else:
        grade='B'
    l.append(grade)
df['GRADES']=l
df
```

```python
x=df[['TEST-1 MARKS']]
y=df['TOTAL']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
